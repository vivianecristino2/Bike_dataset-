### 3. Using the date column, create new columns for: year, month, day of the week and hour of the day

```python
# Convert 'datetime' column to pandas datetime format
bike_dataset['datetime'] = pd.to_datetime(bike_dataset['datetime'])

# Extract year, month, day of the week, and hour
bike_dataset['year'] = bike_dataset['datetime'].dt.year
bike_dataset['month'] = bike_dataset['datetime'].dt.month
bike_dataset['day_of_week'] = bike_dataset['datetime'].dt.day_name()
bike_dataset['hour'] = bike_dataset['datetime'].dt.hour

# Display the updated DataFrame
bike_dataset.head()
```

## 4. Rename the values in the season column to spring, summer, fall and winter.

```python
# Mapeando os valores da coluna 'season' para nomes mais descritivos
bike_dataset['season'] = bike_dataset['season'].replace({
    1: 'spring',
    2: 'summer',
    3: 'fall',
    4: 'winter'
})

# Exibindo as primeiras linhas do DataFrame para verificar
print(bike_dataset.head())
```

### 5. Calculate the total number of casual and registered bikes rented in the years 2011 and 2012

```python
# Filter the dataset for the years 2011 and 2012
filtered_bike_dataset = bike_dataset[(bike_dataset['year'] == 2011) | (bike_dataset['year'] == 2012)]

# Calculate the total number of casual bike rentals
total_casual = filtered_bike_dataset['casual'].sum()

# Calculate the total number of registered bike rentals
total_registered = filtered_bike_dataset['registered'].sum()

# Display the results
print(f"Total Casual Rentals: {total_casual:,}")
print(f"Total Registered Rentals: {total_registered:,}")

```

### 6. Calculate the mean of the hourly total rentals count by season. Which season has the highest mean ?

```python
# Calculate the mean of the hourly total rentals count by season
mean_rentals_by_season = bike_dataset.groupby('season')['count'].mean()

# Determine the season with the highest mean
highest_mean_season = mean_rentals_by_season.idxmax()
highest_mean_value = mean_rentals_by_season.max()

# Display the results
mean_rentals_by_season, highest_mean_season, highest_mean_value
```

```python
# Prepare the results in a DataFrame for better visualization
mean_rentals_table = mean_rentals_by_season.reset_index()
mean_rentals_table.columns = ['Season', 'Mean Rentals']

# Highlight the season with the highest mean rentals
mean_rentals_table['Highlight'] = mean_rentals_table['Season'] == highest_mean_season

# Print the results for better clarity
print(mean_rentals_table)

```

### 7. Are more bikes rented by registered users on working or non-working days ? Does the answer differ for non-registered users ? Is the answer the same for both years ?

```python
# Analyze bike rentals by registered and casual users for working and non-working days
rentals_by_workingday = bike_dataset.groupby(['workingday', 'year'])[['registered', 'casual']].sum().reset_index()

# Add descriptive labels for working and non-working days
rentals_by_workingday['workingday'] = rentals_by_workingday['workingday'].replace({0: 'Non-Working Day', 1: 'Working Day'})

# Display the results in a structured format
rentals_by_workingday
```

### 8. Which months in the year 2011 have the highest and the lowest total number of bikes rented ? Repeat for the year 2012.

```python
# Calculate the total number of bikes rented per month for each year
rentals_by_month = bike_dataset.groupby(['year', 'month'])['count'].sum().reset_index()

# Find the months with the highest and lowest rentals for each year
highest_2011 = rentals_by_month[rentals_by_month['year'] == 2011].sort_values(by='count', ascending=False).iloc[0]
lowest_2011 = rentals_by_month[rentals_by_month['year'] == 2011].sort_values(by='count').iloc[0]

highest_2012 = rentals_by_month[rentals_by_month['year'] == 2012].sort_values(by='count', ascending=False).iloc[0]
lowest_2012 = rentals_by_month[rentals_by_month['year'] == 2012].sort_values(by='count').iloc[0]

# Results
results_2011 = {
    "Highest Month": highest_2011['month'],
    "Highest Rentals": highest_2011['count'],
    "Lowest Month": lowest_2011['month'],
    "Lowest Rentals": lowest_2011['count']
}

results_2012 = {
    "Highest Month": highest_2012['month'],
    "Highest Rentals": highest_2012['count'],
    "Lowest Month": lowest_2012['month'],
    "Lowest Rentals": lowest_2012['count']
}

results_2011, results_2012
```

### 9. Which type of weather have the highest and lowest mean of the hourly total rentals count ?

```python
# Calculate the mean of hourly total rentals count by weather type
weather_mean_rentals = bike_dataset.groupby('weather')['count'].mean().reset_index()

# Find the weather type with the highest and lowest mean rentals
highest_weather = weather_mean_rentals.sort_values(by='count', ascending=False).iloc[0]
lowest_weather = weather_mean_rentals.sort_values(by='count').iloc[0]

# Prepare the results
results_weather = pd.DataFrame({
    "Metric": ["Highest Mean Rentals", "Lowest Mean Rentals"],
    "Weather Type": [highest_weather['weather'], lowest_weather['weather']],
    "Mean Rentals": [highest_weather['count'], lowest_weather['count']]
})

# Display the results
print(results_weather)
```

### 10. Calculate the correlation between the hourly total rentals count and all the numerical columns in the dataset. Which column has the highest correlation with the total rentals count ?

```python
# Calculate correlations of all numerical columns with the 'count' column
correlations = bike_dataset.corr(numeric_only=True)['count'].drop('count').sort_values(ascending=False)

# Find the highest correlation
highest_corr_column = correlations.idxmax()
highest_corr_value = correlations.max()

# Display the results
print("Correlations with 'count':")
print(correlations)
print(f"\nHighest correlation is with '{highest_corr_column}' (Correlation: {highest_corr_value:.4f})")
```

### 11. Create a new categorical column called day_period, which can take four possible values: night, morning, afternoon and evening. These values correspond to the following binning of the hour column: 0-6: night, 6-12: morning, 12-6: afternoon, 6-24:evening.

```python
# Create a new categorical column 'day_period' based on the 'hour' column
def assign_day_period(hour):
    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'

bike_dataset['day_period'] = bike_dataset['hour'].apply(assign_day_period)

# Display the first few rows to confirm the new column
bike_dataset[['hour', 'day_period']].head()
```

### 12. Generate a pivot table for the mean of the hourly total rentals count, with the index set to the day period and the column set to the working day column. What can you observe from the table?

```python
# Generate a pivot table for the mean of the hourly total rentals count
pivot_table = bike_dataset.pivot_table(
    values='count',
    index='day_period',
    columns='workingday',
    aggfunc='mean'
)

# Rename columns for better clarity
pivot_table.columns = ['Non-Working Day', 'Working Day']

# Display the pivot table
pivot_table
```

## 5. Plot the the mean of the hourly total rentals count for the different months for both years separately in a multi-panel figure.
